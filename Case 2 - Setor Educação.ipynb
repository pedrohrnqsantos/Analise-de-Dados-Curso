```python
# lendo as base de dados

base_dados = pd.read_csv('StudentsPerformance+(1).csv')
base_dados.head(5)

# dimens√µes

print('Linhas: {}, Colunas: {}'.format(base_dados.shape[0], base_dados.shape[1]))
```

```python
# segmentando por genero

print(base_dados['gender'].value_counts())
print(base_dados['gender'].value_counts(normalize= True)*100)
```

```python
# segmentando por etnia

print(base_dados['race/ethnicity'].value_counts())
print(base_dados['race/ethnicity'].value_counts(normalize= True)*100)
```

```python
# segmentando por educacional dos pais

print(base_dados['parental level of education'].value_counts())
print(base_dados['parental level of education'].value_counts(normalize= True)*100)
```

```python
# segmentando por op√ß√£o de lanche ( normal ou reduzido/gratuito )

print(base_dados['lunch'].value_counts())
print(base_dados['lunch'].value_counts(normalize= True)*100)
```

```python
# segmentando por alunos que fizeram curso preparatorio

print(base_dados['test preparation course'].value_counts())
print(base_dados['test preparation course'].value_counts(normalize= True)*100)
```

```python
# analise do teste de matematica por sexo

plt.figure(figsize= (20,10))
sns.boxplot(data= base_dados, x='math score', y='gender');
```

```python
# analise do teste de leitura por sexo

plt.figure(figsize= (20,10))
sns.boxplot(data= base_dados, x='reading score', y='gender');
```

```python
# analise do teste de escrita por sexo

plt.figure(figsize= (20,10))
sns.boxplot(data= base_dados, x='writing score', y='gender');
```

```python
base_dados.groupby(by= ['gender']).describe()['math score'].reset_index()


# outlier da base feminina acabou puxando demais a m√©dia.
# mas verificando pelos quartis ainda sim ficaria abaixo da m√©dia masculina
# obs: no teste de matematica
```

```python
# analise grafica por etinia pairplot

plt.figure(figsize= (20,10))
sns.pairplot(base_dados, hue='race/ethnicity')
```

```python
# analise gr√°fica por etnia boxplot

plt.figure(figsize= (20,15))
sns.boxplot(data= base_dados, x= 'math score', y='race/ethnicity');
```

```python
# analise gr√°fica por educacional dos pais boxplot

plt.figure(figsize= (20,15))
sns.boxplot(data= base_dados, x= 'math score', y='parental level of education');
```

```python
base_dados.groupby(by= ['parental level of education']).describe()['math score'].reset_index()

# os filhos que tem pais apenas com ensino medio, tiveram a menor m√©dia em matematica
# em consequencia os pais que possuem mestrado, seus filho tiveram uma media.
# obseva√ß√£o √© que todos est√£o na casa de 60-70 de media
# os outliers s√£o aparentes para o menor nivel educacional
```

```python
# analise grafica de quem fez o teste preparatorio ou nao

plt.figure(figsize= (20,15))
sns.boxplot(data= base_dados, x='math score',
            y= 'test preparation course');
```

```python
base_dados.groupby(by= ['test preparation course']).describe()['math score'].reset_index()

# na media quem fez o curso se saiu melhor
# a varia√ß√£o foi de +-5 pts
# os outliers minimos de quem nao fez foram maiores e aparentes.
```
