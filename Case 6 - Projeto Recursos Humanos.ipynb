```python
# importando as librarys

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# lendo a base de dados

base_dados = pd.read_csv('Salary_Data.csv')
base_dados.head()
```

```python
# dimensoes da base de dados

print('Linhas: {}, Colunas: {}'.format(base_dados.shape[0], base_dados.shape[1]))
```

```python
# renomeando as colunas 

base_dados.rename( columns= {
    'YearsExperience' : 'Experiência',
    'Salary' : 'Renda'
}, inplace= True)
```

```python
# informações da base de dados

base_dados.info()
```

```python
# campos nulos

print(base_dados.isnull().sum())

# gráfico

plt.figure(figsize= (15,7))
sns.heatmap(base_dados.isnull())
```

```python
# estatistica

base_dados.describe()
```

```python
# grafico analise renda

plt.figure(figsize= (20,10), facecolor= 'gray')
sns.kdeplot(base_dados['Renda'],
            color= 'green');
```

```python
# grafico analise experiencia

plt.figure(figsize= (20,10), facecolor= 'gray')
plt.title('Experência', fontsize= 25)
sns.kdeplot(base_dados['Experiência'],
            color= 'green');
```

```python

plt.figure(figsize= (20,10))
plt.title('Experência x Renda', fontsize= 20)
sns.lineplot(data= base_dados, x='Experiência', y= 'Renda',
             estimator= 'mean');
```
