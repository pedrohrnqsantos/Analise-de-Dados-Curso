```python
# importando as librarys

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# lendo a base de dados

base_dados = pd.read_excel('Dados_Empresas_Energia.xlsx')

# registros iniciais

base_dados.head()
```

```python
# dimensões da base de dados

print(f'Linhas: {base_dados.shape[0]},',
      f'Colunas: {base_dados.shape[1]}')
```

```python
# informações dos tipos de dados

base_dados.info()
```

```python
# estatisticas

base_dados.describe()
```

```python
# setar o index

base_dados.set_index('Data', inplace= True)
base_dados.head(3)
```

```python
# Gráfico

# estilo grafico
plt.style.use('seaborn-darkgrid')

# tamanho
plt.figure(figsize= (16,7))

# titulo
plt.title('Análise Ações Empresas de Energia',
          loc= 'left',
          fontsize= 18,
          fontweight= 600);

# Plot Petrobras
plt.plot(base_dados.index, base_dados.Petrobras, color= '#008C4A', linewidth= 4, alpha= 0.7)
plt.text(base_dados.index[-1], base_dados['Petrobras'].tail(1),
         'Petrobras',color= '#008C4A', size= 'large', horizontalalignment= 'left')

# Plot de todas as colunas
for coluna in base_dados.columns[1:]:
    plt.plot(base_dados.index, base_dados[coluna], color= 'gray', linewidth= 2, alpha= 0.7)
    plt.text(base_dados.index[-1], base_dados[coluna].tail(1),
         coluna,color= 'gray', size= 'large', horizontalalignment= 'left')

# labels
plt.xlabel('Período')
plt.ylabel('Preço de Fechamento (R$)');
```
