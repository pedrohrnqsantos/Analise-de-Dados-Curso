```python
# importando as librarys

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# lendo a base de dados

base_dados = pd.read_excel('Dados_Pib.xlsx')
base_dados.head()
```

```python
# dimensoes

print(f'Linhas: {base_dados.shape[0]},',
      f'Colunas: {base_dados.shape[1]}')
```

```python
# informações da base de dados

base_dados.info()
```

```python
# campos nulos

base_dados.isnull().sum()

# grafico

plt.figure(figsize= (12,6))
sns.heatmap(base_dados.isnull())
```

```python
# estatisticas

base_dados.describe()
```

```python
# Sistem de Grids

# cor de fundo
cor_de_fundo = '#F5F5F5'

#incluir cor de fundo


# criar sistema de grid
grid_graficos = sns.FacetGrid(base_dados, col= 'Territorialidades',
                hue= 'Territorialidades', col_wrap= 4)

# adicinar um grafico de linhas em cada grafico
grid_graficos = grid_graficos.map(plt.plot, 'Ano', 'PIB per capita')

# adicionar sombra
grid_graficos = grid_graficos.map(plt.fill_between, 'Ano', 'PIB per capita', alpha= 0.3).set_titles('{col_name}')

# adicionando titulo
grid_graficos = grid_graficos.fig.suptitle('Evolução da Renda per Capita por Estado', fontsize= 18)

# ajuste titulo
plt.subplots_adjust(top= 0.95)
```
